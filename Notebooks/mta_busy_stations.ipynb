```python
busyEntries = df.groupby(['station']).deltaEntry.sum().sort_values(ascending=False).head(30)
busyEntries
```

```python
busyExits = df.groupby(['station']).deltaExit.sum().sort_values(ascending=False).head(30)
busyExits
```

```python
busyStations = (busyEntries+busyExits).keys()
busyStations
```

```python
busy_station_df = df[df['station'].isin(busyStations)]
```
